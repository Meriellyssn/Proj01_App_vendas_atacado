#Exercicio 01

```python
print('Bem-vindo(a) a Loja da Meirielli S. Sousa do N.')
valor_produto = float(input('Digite o valor do produto:R$ '))
qtd_produto = int(input('Digite a quantidade desejada: '))
desc_produto = 0
#Descontos a serem aplicados
if qtd_produto <= 200:
  desc_produto = 0.00
elif qtd_produto >= 200 and qtd_produto < 1000:
  desc_produto = 0.05
elif qtd_produto >= 1000 and qtd_produto < 2000:
  desc_produto = 0.10
else:
  desc_produto = 0.15
#Valores com e sem descontos
total_sem_desc = valor_produto * qtd_produto
print("O valor total SEM desconto é de:R$ {:.2f}".format(total_sem_desc))
total_com_desc = total_sem_desc - total_sem_desc * desc_produto
print('O valor total COM desconto é de:R$ {:.2f}'.format(total_com_desc))
```

#Exercicio 03

```python
#Inicio da função cachorro_peso()
def cachorro_peso():
  print('-------------------------- Menu 1 de 3 - Peso kg Cachorro --------------------------')
  while True:
    try:
      peso = float(input('\nDigite o peso do cachorro(kg): '))
      if (peso < 3):
        return 40
      elif (peso >=3) and (peso < 10):
        return 50
      elif (peso >=10) and (peso < 30):
        return 60
      elif (peso >=3) and (peso <50):
        return 70
      else:
        print('Não aceitamos cachorros tão grandes. Por favor digite novamente.')
    except ValueError: #Caso o usuário digite letra ou um valor com virgula em vez de ponto.
      print('Peso inválido. Por favor, digite novamente.')
      continue #Retorna para a pergunta inicial
#Fim da função cachorro_peso()

#Inicio da função cachorro_pelo()
def cachorro_pelo():
  print('-------------------------- Menu 2 de 3 - Pelo Cachorro --------------------------')
  while True:
    pelo = input('\nEntre com o pelo do cachorro: \n' +
                 'c - Pelo Curto \n' +
                 'm - Pelo Médio \n' +
                 'l - Pelo Longo \n' +
                 '>>: ')
    pelo = pelo.lower()
    pelo = pelo.strip()
    if pelo =='c':
      return 1
    elif pelo == 'm':
      return 1.5
    elif pelo == 'l':
      return 2
    else:
      print('Opção inválida. Digite a opção desejada novamente.')
      continue #Voltar para o inicio
#Fim da função cachorro_pelo()

#Inicio da função cachorro_extra()
def cachorro_extra():
  print('-------------------------- Menu 3 de 3 - Extra --------------------------')
  acumulador = 0
  while True:
    adicional = input('\nDeseja adicionar mais algum serviço? \n' +
                      '1 - Corte de Unhas - R$ 10,00 \n' +
                      '2 - Escovar Dentes - R$ 12,00' +
                      '3 - Limpeza de Orelhas - R$ 15,00 \n' +
                      '0 - Não desejo mais nada \n' +
                      '>>: ')
    if adicional == '1':
      acumulador = acumulador + 10
      continue
    elif adicional =='2':
      acumulador = acumulador + 12
      continue
    elif adicional == '3':
      acumulador = acumulador + 15
      continue
    elif adicional == '0':
      return acumulador
    else:
      print('Opção inválida. Por favor digite novamente.')
#Fim da função cachorro_extra()

#Inicio do Main
print('--------------- Bem-vindo(a) ao Pet Shop da Meirielli S. Sousa do N. ---------------')
peso = cachorro_peso()
pelo = cachorro_pelo()
adicional = cachorro_extra()
total = (peso * pelo) + adicional
print('Total a pagar(R$): {:.2f} (peso: {:.2f} * pelo: {:.2f} + adicional(is): {:.2f})'.format(total, peso, pelo, adicional))
print('                           Obrigado pela preferência!')
print('                                 Volte sempre!')
```

#Exercicio 4

```python
#----------- Inicio das Variáveis Globais -----------
lista_colaboradores = []
id_global = 0
#----------- Fim das Variáveis Globais -----------

#----------- Inicio de cadastrar_colaborador(id) -----------
def cadastrar_colaborador(id):
  print('*********************************************************************')
  print('--------- Bem-vindo(a) ao Menu de Cadastro de Colaboradores ---------\n')
  print('Id do Colaborador: {}'.format(id_global))
  nome = input('Por favor entre com o NOME: ')
  setor = input('Por favor entre com o SETOR: ')
  pagamento = float(input('Por favor entre com o PAGAMENTO(R$): '))
  dicionario_colaborador = {'id'        : id,
                            'nome'      : nome,
                            'setor'     : setor,
                            'pagamento' : pagamento}
  lista_colaboradores.append(dicionario_colaborador.copy())
  print('\n---------------- Colaborador Cadastrado com Sucesso! ----------------')
  print('*********************************************************************')
#----------- Fim de cadastrar_colaborador(id) -----------

#----------- Inicio de consultar_colaborador() -----------
def consultar_colaborador():
  print('*********************************************************************')
  while True:
    print('--------- Bem-vindo(a) ao Menu de Consultar Colaborador(es) ---------')
    opcao_consultar = input('\nEscolha a opção desejada:\n'+
                            '1 - Consultar TODOS os Colaboradores \n'+
                            '2 - Consultar Colaborador por ID \n'+
                            '3 - Consultar Colaborador(es) por SETOR \n'+
                            '4 - Retornar \n'+
                            '>>:' )
    if opcao_consultar == '1': #Opção para consultar todos od colaboradores.
     print('*********************************************************************')
     print('Você escolheu a opção Consultar TODOS os Colaboradores')
     for colaborador in lista_colaboradores: #Colaborador vai varrer toda a lista de lista_colaboradores
       print('--------------------------------------------------------------------')
       for key, value in colaborador.items(): #Varrer todos os conjuntos chave e valor do dicionario colaborador
        print('{}: {}'.format(key, value))
       print('--------------------------------------------------------------------')
    elif opcao_consultar == '2': #Opção para consultar colaborador por id
     print('*********************************************************************')
     print('Você escolheu a opção Consultar Colaborador por ID \n')
     id_desejado = int(input('Digite o ID do Colaborador: '))
     for colaborador in lista_colaboradores:
      if colaborador['id'] == id_desejado: #O valor do campo id desse dicionario é igual ao valor desejado?
        print('--------------------------------------------------------------------')
        for key, value in colaborador.items(): #Varrer todos os conjuntos chave e valor do dicionario colaborador
          print('{}: {}'.format(key, value))
        print('--------------------------------------------------------------------')
    elif opcao_consultar == '3': #Opção consultar colaboradores por setor
     print('********************************************************************')
     print('Você escolheu a opção Consultar Colaborador(es) por SETOR \n')
     setor_desejado = input('Digite o SETOR do(s) Colaborador(es): ')
     for colaborador in lista_colaboradores:
      if colaborador['setor'] == setor_desejado: #O valor do campo id desse dicionario é igual ao valor desejado?
        print('--------------------------------------------------------------------')
        for key, value in colaborador.items(): #Varrer todos os conjuntos chave e valor do dicionario colaborador
          print('{}: {}'.format(key, value))
        print('--------------------------------------------------------------------')
    elif opcao_consultar == '4':
     return #Sair da opçao consultar colaboradore(es) e volt para o Main
    else:
      print('Opção Inválida. Tente Novamente.')
      continue #Volta para o inicio do laço
#----------- Fim de consultar_colaborador() -----------

#----------- Inicio de remover_colaborador() -----------
def remover_colaborador():
  print('********************************************************************')
  print('------------ Bem-vindo(a) ao Menu de Remover Colaborador ----------- \n')
  id_desejado = int(input('Digite o ID do colaborador a ser removido: '))
  for colaborador in lista_colaboradores:
    if colaborador['id'] == id_desejado:
      lista_colaboradores.remove(colaborador)
      print('\n------------------ Colaborador removido com Sucesso! ---------------')

#----------- Fim de remover_colaborador() -----------

#----------- Inicio do Main -----------
print('Bem-vindo(a) ao Controle de Colaboradores da Meirielli S. Sousa do N.')
print('*********************************************************************')
while True:
  print('-------------------------- Menu Principal ---------------------------')
  opcao_principal = input('\nEscolha a opção desejada:\n'+
                          '1 - Cadastrar Colaborador \n'+
                          '2 - Consultar Colaborador(es) \n'+
                          '3 - Remover Colaborador \n'+
                          '4 - Sair \n'+
                          '>>:' )
  if opcao_principal == '1':
    id_global = id_global + 1
    cadastrar_colaborador(id_global)
  elif opcao_principal == '2':
    consultar_colaborador()
  elif opcao_principal == '3':
    remover_colaborador()
  elif opcao_principal == '4':
    print('\n Programa Encerrado...')
    break #Encerra o laço principal e o programa acaba
  else:
    print('Opção Inválida. Tente Novamente.')
    continue #Volta para o inicio do laço

#----------- Fim do Main -----------
```
